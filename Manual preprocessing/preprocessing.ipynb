### Function called as df.fillna(value)

```python

def replacer(df):
    cat = list(df.columns[df.dtypes=='object'])
    for i in df.columns:
        if i in cat:
            md = df[i].mode()[0]
            df[i] = df[i].fillna(md)
        else:
            mn = df[i].mean()
            df[i] = df[i].fillna(mn)
    print('Missing values replaced')
```

### Apply one hot ecoding on categorical features

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(handle_unknown='ignore',
                    sparse_output=False).set_output(transform='pandas')
X_cat_pre = ohe.fit_transform(X_cat)
X_cat_pre.head()
```

```python
X_cat_pre.to_csv("ohe.csv", index=False)
```

### Apply standard scaler on numerical features

```python

from sklearn.preprocessing import StandardScaler
```

```python

scaler = StandardScaler().set_output(transform='pandas')
X_con_pre = scaler.fit_transform(X_con)
X_con_pre.head()
```

### Combine both categorical and continous preprocessed data

### Apply Train Test split

2.20% of data will be Test : These values are hidden from model

### Evaluate the model
1.Mean Squared Error

2.Root Mean Squared Error

3.Mean Absolute Error

4.Mean Absolute Percentage Error

5.R2 Score


```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, \
                            mean_absolute_percentage_error, r2_score
```

```python
def evaluate_model(model, x, y):
    # Predict the results from model
    ypred = model.predict(x)
    # Get all the evaluation metrics
    mse = mean_squared_error(y, ypred)
    rmse = mse**(1/2)
    mae = mean_absolute_error(y, ypred)
    mape = mean_absolute_percentage_error(y, ypred)
    r2 = r2_score(y, ypred)
    # Print all the results
    print(f'Mean Squared Error : {mse:.2f}')
    print(f'Root Mean Squared Error : {rmse:.2f}')
    print(f'Mean Absolute error : {mae:.2f}')
    print(f'Mean Absolute Percentage Error : {mape:.4f}')
    print(f'R2 Score : {r2:.4f}')
```

```python
# Training Data results
evaluate_model(model, xtrain, ytrain)
```

```python

# Test data results
evaluate_model(model, xtest, ytest)
```


### By seeing above results model has test r2 of 0.9331 > 0.8 Then it is a good model
